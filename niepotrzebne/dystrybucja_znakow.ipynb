```python
import matplotlib.pyplot as plt
from collections import Counter

def analyze_character_distribution(text, char_set):
    """
    Analyzes the frequency of characters in the given text from a specified set of characters.

    Args:
    text (str): The text to analyze.
    char_set (set): A set of characters to consider in the analysis.

    Returns:
    dict: A dictionary with characters as keys and their frequencies as values.
    """
    # Filter the text to include only characters in char_set
    filtered_text = [char for char in text if char in char_set]
    
    # Count the occurrences of each character
    char_count = Counter(filtered_text)
    
    return char_count

def plot_character_distribution(char_count):
    """
    Plots the distribution of characters.

    Args:
    char_count (dict): A dictionary with characters as keys and their frequencies as values.
    """
    # Sort the dictionary by character for consistent plotting
    sorted_chars = sorted(char_count.items())
    characters, counts = zip(*sorted_chars)

    plt.figure(figsize=(10, 6))
    plt.bar(characters, counts)
    plt.xlabel("Characters")
    plt.ylabel("Frequency")
    plt.title("Character Distribution in Pan Tadeusz (Lowercase)")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example usage
with open("pan-tadeusz.txt", encoding="utf-8") as file:
    pan_tadeusz_text = file.read()

# Convert text to lowercase
pan_tadeusz_text_lower = pan_tadeusz_text.lower()

# Define the set of characters to analyze (e.g., all letters and punctuation)
chars = sorted(list(set(pan_tadeusz_text)))

# Analyze the lowercase text
char_count_lower = analyze_character_distribution(pan_tadeusz_text_lower, chars)

# Plot the distribution
plot_character_distribution(char_count_lower)
```

```python
# Remove unnecessary whitespace from the text
pan_tadeusz_cleaned = " ".join(pan_tadeusz_text_lower.split())


# Analyze the cleaned text for character distribution
char_count_cleaned = analyze_character_distribution(pan_tadeusz_cleaned, chars)

# Plot the distribution for cleaned text
plot_character_distribution(char_count_cleaned)

```

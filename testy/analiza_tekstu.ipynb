```python
def count_total_characters(text):
    """
    Funkcja liczy łączną liczbę znaków w tekście, pomijając spacje.
    
    Args:
        text (str): Tekst wejściowy.
        
    Returns:
        int: Łączna liczba znaków (bez spacji).
    """
    text_no_spaces = text.replace(" ", "")  # Usuń spacje
    return len(text_no_spaces)  # Policz długość tekstu bez spacji

# Wczytaj tekst
with open('pan-tadeusz.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# Obliczenie sumy wszystkich znaków (bez spacji)
total_characters = count_total_characters(text)

# Wyświetlenie wyniku
print(f"Łączna liczba znaków (bez spacji): {total_characters}")
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Wczytaj tekst z pliku
with open('pan-tadeusz-oczyszczony.txt', 'r', encoding='utf-8') as f:
    text = f.read().lower()  # Zmniejsz wszystkie litery na małe

# Usuń spacje i zlicz wystąpienia znaków
text_no_spaces = text.replace(" ", "").replace("\n", "")  # Usuń spacje i znaki nowej linii
char_counts = Counter(text_no_spaces)

# Filtrowanie znaków występujących rzadziej niż 3 razy
filtered_counts = {char: count for char, count in char_counts.items() if count >= 3}

# Przygotowanie danych do wykresu
chars_filtered = list(filtered_counts.keys())
counts_filtered = list(filtered_counts.values())

# Sortowanie znaków według ich częstotliwości
sorted_indices = sorted(range(len(counts_filtered)), key=lambda k: counts_filtered[k], reverse=True)
chars_sorted = [chars_filtered[i] for i in sorted_indices]
counts_sorted = [counts_filtered[i] for i in sorted_indices]

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
plt.bar(chars_sorted, counts_sorted, alpha=0.75, color='blue')
plt.title("Dystrybucja znaków w tekście (bez spacji, małe litery, filtrowane)", fontsize=14)
plt.xlabel("Znaki", fontsize=12)
plt.ylabel("Częstotliwość", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

# Wczytaj tekst
with open('pan-tadeusz.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# Podstawowe statystyki
total_chars = len(text)  # Wszystkie znaki (ze spacjami)
text_no_spaces = text.replace(" ", "").replace("\n", "")  # Usuwamy spacje i znaki nowej linii
chars_no_spaces = len(text_no_spaces)  # Liczba znaków bez spacji

lines = text.splitlines()  # Wszystkie linie
non_empty_lines = [line for line in lines if line.strip()]  # Linie niepuste

# Tokeny jako znaki
unique_chars = len(set(text_no_spaces))  # Liczba unikalnych znaków
char_counts = Counter(text_no_spaces)  # Liczba wystąpień każdego znaku

# Statystyki dotyczące linii
avg_chars_per_line = sum(len(line) for line in non_empty_lines) / len(non_empty_lines) if non_empty_lines else 0

# Najczęstsze znaki
most_common_chars = char_counts.most_common(10)

# Wyniki
print(f"Całkowita liczba znaków (ze spacjami): {total_chars}")
print(f"Całkowita liczba znaków (bez spacji): {chars_no_spaces}")
print(f"Liczba linii: {len(lines)}")
print(f"Liczba niepustych linii: {len(non_empty_lines)}")
print(f"Liczba unikalnych znaków: {unique_chars}")
print(f"Średnia liczba znaków na linię (bez pustych): {avg_chars_per_line:.2f}")
print(f"Najczęstsze znaki: {most_common_chars}")

```

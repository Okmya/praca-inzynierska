```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie danych z pliku CSV
file_path = "wynik_z_iteracja.csv"  # Zamie na rzeczywist ciek pliku
data = pd.read_csv(file_path)

# Wybranie iteracji do oznaczenia na wykresie
highlight_iterations = [1, 1500, 3000, 4500, 6000]

# Filtracja danych dla iteracji do oznaczenia
filtered_data = data[data['iteracja'].isin(highlight_iterations)]

# Wycignicie danych do wykresu
iterations = data['iteracja']
train_perplexity = data['final_train_perplexity']
val_perplexity = data['final_val_perplexity']

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
plt.plot(iterations, train_perplexity, label='Nieokrelono na zbiorze treningowym', color='orange', linewidth=1)
plt.plot(iterations, val_perplexity, label='Nieokrelono na zbiorze walidacyjnym', color='green', linewidth=1)

# Dodanie duych grubych kropek tylko dla zbioru walidacyjnego
plt.scatter(filtered_data['iteracja'], filtered_data['final_val_perplexity'], color='blue', s=100, label='Wybrane iteracje (walidacja)', marker='o')

# Dodanie tytuw osi i legendy
plt.xlabel('Iteracja', fontsize=12)
plt.ylabel('Nieokrelono', fontsize=12)
plt.legend()
plt.grid(True)

# Wywietlenie wykresu
plt.title('Nieokrelono na zbiorze treningowym i walidacyjnym', fontsize=14)
plt.show()
```

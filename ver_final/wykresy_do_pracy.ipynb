```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie danych z pliku CSV
file_path = "wynik_szekspir_model_koncowy.csv"  # Zamień na rzeczywistą ścieżkę pliku
data = pd.read_csv(file_path)

# Wybranie iteracji do oznaczenia na wykresie
highlight_iterations = [1, 1500, 3000, 4500, 6000]

# Filtracja danych dla iteracji do oznaczenia
filtered_data = data[data['iteracja'].isin(highlight_iterations)]

# Wyciągnięcie danych do wykresu
iterations = data['iteracja']
train_perplexity = data['final_train_perplexity']
val_perplexity = data['final_val_perplexity']

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
plt.plot(iterations, train_perplexity, label='Nieokreśloność na zbiorze treningowym', color='darkorange', linewidth=1)
plt.plot(iterations, val_perplexity, label='Nieokreśloność na zbiorze walidacyjnym', color='gold', linewidth=1)

# Dodanie dużych grubych kropek tylko dla zbioru walidacyjnego
plt.scatter(filtered_data['iteracja'], filtered_data['final_val_perplexity'], color='blue', s=100, label='Wybrane iteracje (walidacja)', marker='o')

# Dodanie tytułów osi i legendy
plt.xlabel('Iteracja', fontsize=12)
plt.ylabel('Nieokreśloność', fontsize=12)
plt.legend()
#plt.grid(True)

# Wyświetlenie wykresu
plt.tight_layout()
plt.show()
```

```python
# Extract the relevant columns for plotting
iterations = data['iteracja']
train_loss = data['final_train_loss']
val_loss = data['final_val_loss']
train_perplexity = data['final_train_perplexity']
val_perplexity = data['final_val_perplexity']

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the loss values on the first subplot
axes[0].plot(iterations, train_loss, label='Funkcja straty na zbiorze treningowym', marker='o', color = 'gold')
axes[0].plot(iterations, val_loss, label='Funkcja straty na zbiorze walidacyjnym', marker='o', color = 'darkorange')
#axes[0].set_title('Function Loss vs Iteration')
axes[0].set_xlabel('Iteracja')
axes[0].set_ylabel('funkcja straty')
axes[0].legend()
axes[0].grid()

# Plot the perplexity values on the second subplot
axes[1].plot(iterations, train_perplexity, label='Perplexity na zbiorze treningowym', marker='o', color = 'gold')
axes[1].plot(iterations, val_perplexity, label='Perplexity na zbiorze walidacyjnym', marker='o',color = 'darkorange')
axes[1].set_xlabel('Iteracja')
axes[1].set_ylabel('Perplexity')
axes[1].legend()
axes[1].grid()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
